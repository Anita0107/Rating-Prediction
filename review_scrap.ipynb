```python
stars=[]
urls=[]
xls=[]
short_review=[]
complete_review=[]

# Fetching url for products
for i in driver.find_elements_by_xpath("//a[@class='s1Q9rs']"):
    xls.append(i.get_attribute("href"))

print("Total products to scrape :",len(xls))
    
for cnt_prd in range(len(xls)): # for loop to fetch product review
    count1 = True # setting count1 to true. It means there are more reviews to scrape in next page.
    driver.get(xls[cnt_prd])
    time.sleep(2)
    
    try:
        search_btn=driver.find_element(By.XPATH,"//div[@class='_1AtVbE col-12-12']//div[@class='_3UAT2v _16PBlm']/span") # click on all review
        search_btn.click()
        time.sleep(1)
        try:
         #scraping 10 pages url
            url_1 = driver.find_elements_by_xpath("//a[@class='ge-49M _2Kfbh8']")
            for i in url_1:
                urls.append(i.get_attribute('href'))
            url_2 = driver.find_elements_by_xpath("//a[@class='ge-49M']")
            for i in url_2:
                urls.append(i.get_attribute('href'))
                time.sleep(4)


            for i in urls:
                driver.get(i)
            #for scrapping the number of stars
                for j in driver.find_elements_by_xpath("//div[@class='col _2wzgFH K0kLPL']/div[1]/div[1]"):
                    stars.append(j.text)
         
                for l in driver.find_elements_by_xpath("//div[@class='t-ZTKy']/div/div"):
                    complete_review.append(l.text)
        
        except:
            count1 = False # Setting count1 to false. It means there are no review to scrape.
            
    except:
        pass
    
    print("Completed scraping for product",cnt_prd+1) # status for scraping
    
review_list6 = pd.DataFrame({})
review_list6['Rating'] = stars[0:]
review_list6['Review'] = complete_review[0:]

review_list6
```

```python
#Amazon 1
driver.get("https://www.amazon.in/")

time.sleep(2)

search_job=driver.find_element(By.XPATH,"//input[@class='nav-input nav-progressive-attribute']")  #search bar
search_job.send_keys("laptop")

search_btn=driver.find_element(By.XPATH,"//div[@class='nav-right']//input[@class='nav-input nav-progressive-attribute']") #search button
search_btn.click()

time.sleep(2)

urls=[]
rating_info = []
review_info = []

# Fetching url for products
for i in driver.find_elements_by_xpath("//h2//a[@class='a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal']"):
    urls.append(i.get_attribute("href"))

print("Total products to scrape :",len(urls))
    
for cnt_prd in range(len(urls)): # for loop to fetch product review
    count1 = True # setting count1 to true. It means there are more reviews to scrape in next page.
    driver.get(urls[cnt_prd])
    time.sleep(2)
    
    try:
        search_btn=driver.find_element(By.XPATH,"//a[@data-hook='see-all-reviews-link-foot']") # click on all review
        search_btn.click()
        time.sleep(1)
        
        while(count1): # while loop till count1 is true. It means countinue the review scarping as there are more reviews in next page.
            try:
                for cnt in range(10): # for loop to fetch review in different pages
                    
                    # Scraping rating
                    info = driver.find_elements(By.XPATH,"//div[@class='a-section celwidget']//div[@class='a-row']//a[@class='a-link-normal']")
                    rating_info.append(info[cnt].get_attribute("title").split('.')[0])
                    
                    # Scraping review
                    info = driver.find_elements(By.XPATH,"//div[@class='a-row a-spacing-small review-data']")
                    review_info.append(info[cnt].text)
                    
                # Clicking next button
                search_btn=driver.find_element(By.XPATH,"//li[@class='a-last']")
                search_btn.click()
                time.sleep(1)
            
            except:
                count1 = False # Setting count1 to false. It means there are no review to scrape.
            
    except:
        pass
    
    print("Completed scraping for product",cnt_prd+1) # status for scraping
    
```

```python
review_list1 = pd.DataFrame({})
review_list1['Rating'] = rating_info[0:]
review_list1['Review'] = review_info[0:]

review_list1
```
